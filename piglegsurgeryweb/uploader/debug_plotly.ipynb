```python
cols = [
    "Needle holder visibility [s]",
    "Forceps visibility [s]",
]

# Normalize by median and prepare hover data
for col in cols:
    normalized_col = col + " normalized"
    df[normalized_col] = df[col] / df[col].median()
    df[normalized_col + " hover"] = df[col].astype(str) + " s"

# Melt the DataFrame for Plotly
df_melted = df.melt(id_vars=[col + " normalized hover" for col in cols], 
                    value_vars=[col + " normalized" for col in cols], 
                    var_name='Measurement', 
                    value_name='Normalized Value')

# Plot
fig = px.box(df_melted, x='Measurement', y='Normalized Value', 
             hover_data=[col + " normalized hover" for col in cols])

# Customize hover text
# fig.update_traces(hovertemplate='%{hovertext}')
# 
fig.show()
```
